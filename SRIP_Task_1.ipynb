#### Importing libraries

```python
import torch
import math
import matplotlib.pyplot as plt
import random
import numpy as np
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

import os    
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

#### Defining class `frechet_dist`
* Contains `func` function which is the PDF of the distribution
* Contains `cdf` function
* Contains `plot` function for visualization 

```python
class frechet_dist():
    def __init__(self,a,s,m):
        self.a=a
        self.s=s
        self.m=m
    
    def func(self, x: torch.Tensor) -> torch.Tensor:
        y = (self.a/self.s) * (((x-self.m)/self.s)**(-1-self.a)) * np.exp(- ((x-self.m)/self.s)**(-self.a) )
        return y
    
    def plot(self,x,y,title,xlabel=None, ylabel=None):
        plt.plot(x,y)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.show()
        
    def cdf(self, x):
        return math.e ** (-1 * ((x-self.m)/self.s) ** (-1 * self.a))
```

#### Creating `dist_` with given values of `a`,`s`,`m`

```python
a=1.0
s=1.0
m=0.0
dist_=frechet_dist(a,s,m)
```

#### Plotting `PDF`, `CDF`, and the inverse of `CDF` of the distribution

```python
x=torch.arange(0.02,10,0.02)
y=dist_.func(x)
dist_.plot(x,y,"PDF","x","probability")

cdf=dist_.cdf(x)
dist_.plot(x,cdf, "CDF", "x")
dist_.plot(cdf,x,"CDF Inverse", ylabel="x")
```

#### Implementing `Inverse CDF Sampling` for `1000` samples



```python
samples=1000
ran=[]

for i in range(samples):
    ran.append(0.9*random.random())

sample=m+s*(-1*np.log(ran)) ** (-1.0/a)

sns.kdeplot(sample)
sns.lineplot(x,y).set(title=r"""$Samples : {}$
$\alpha: 1, s=1, m=0$""".format(samples))
```
